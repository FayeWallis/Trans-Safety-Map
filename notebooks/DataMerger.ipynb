Concatenating our two data sets into a single uniform DataFrame

```python
import pandas as pd
from google.colab import drive
import numpy as np

drive.mount('/content/gdrive')
root_path = 'gdrive/My Drive/Colab Notebooks/data files'


```

```python
pd.options.mode.chained_assignment = None  
df_hrc = pd.read_csv('gdrive/My Drive/Colab Notebooks/TSM Project/data files/hrc_data.csv')
df_hate = pd.read_csv('gdrive/My Drive/Colab Notebooks/TSM Project/data files/hate_crime.csv', low_memory=False)
df2_hate = df_hate[['DATA_YEAR', 'PUB_AGENCY_NAME', 'AGENCY_TYPE_NAME', 'STATE_ABBR', 'BIAS_DESC', 'OFFENSE_NAME']]
df3_hate = df2_hate.loc[(df2_hate['DATA_YEAR'] > 2014) & (df2_hate['BIAS_DESC'] == 'Anti-Transgender')]
df3_hate.rename(columns = {'DATA_YEAR':'Year', 'PUB_AGENCY_NAME':'City', 
                                    'STATE_ABBR':'State', 'OFFENSE_NAME':'Severity'}, inplace = True)
df_fbi = df3_hate[['Year', 'City', 'State', 'Severity']]
severity = ['Murder'] * 213
df_hrc['Severity'] = severity
```

```python
df_summary = df_hrc.append(df_fbi, ignore_index=True)
df_summary.to_csv('gdrive/My Drive/Colab Notebooks/TSM Project/data files/summary_data.csv')
```
