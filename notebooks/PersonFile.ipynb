# Splitting the data into chunks in order to process it in core

```python
genid_target = None
for chunk in pd.read_csv('gdrive/My Drive/Colab Notebooks/TSM Project/tsv files/PD.tsv', sep='\t', chunksize=10000, low_memory=False):
  trans = chunk[(chunk['V3085'] != chunk['V3086']) & (chunk['V3086'] != 8)]
  if len(trans) > 0:
    if genid_target is None:
      genid_target = trans
    else:
      genid_target = pd.concat([genid_target, trans])
      print(genid_target.shape)

print(genid_target.shape)
genid_target.head()
```
