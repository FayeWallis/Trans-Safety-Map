```python
from geopy.extra.rate_limiter import RateLimiter
from geopy.geocoders import Nominatim
from tqdm import tqdm
import pandas as pd
tqdm.pandas()
```

```python
geolocator = Nominatim(user_agent="Trans Safety Map")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
```

```python
geocode('cleveland, ohio')
```

```python
df = pd.read_csv('../database/data/summary_data.csv')
df
```

```python
df['geocode_output'] = df.progress_apply(lambda x: geocode(x['City'] + ', ' + x['State']), axis=1)
```

```python
df['point'] = df.geocode_output.apply(lambda x: f'POINT({x[1][0]} {x[1][1]})' if x else None)
```

```python
df
```

```python
df[['ID', 'City', 'State', 'Year', 'Severity', 'geocode_output', 'point']].to_csv('../database/data/summary_data.csv', index=False)
```
