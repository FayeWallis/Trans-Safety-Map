```python
obituaries = [string_21, string_20, string_19, string_18, 
              string_17, string_16, string_15]
#create lists that correlate to each column of the 2022 dataframe
cities = []
states = []
#create a number of 'years' values to match each obituary chunk
years = [2021] * 56 + [2020] * 37 + [2019] * 24 + [2018] * 28 + [2017] * 26 + [2016] * 21 + [2015] * 21
#gather the obituary chunks into coherent lists
for obit in obituaries:
  for x in obit:
    if len(x) > 2:
      cities.append(x.replace('-', ' '))
    else: 
      states.append(x)
```

```python
#append 2015-2021 dataframes to the original 2022 dataframe
df2 = pd.DataFrame({'City': cities, 'State': states, 'Year': years})
df.append(df2, ignore_index=True)
```
