# Web scraper to get information from the HRC website

```python
def text_scraper(urls):
  for url in urls.keys():
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    #the 2021 page is formatted differently than 2022
    if url == 'https://www.hrc.org/resources/fatal-violence-against-the-transgender-and-gender-non-conforming-community-in-2021':
      paragraphs = soup.find_all('li')
    else:
      paragraphs = soup.find_all('p')
    year = urls.get(url)

    for p in paragraphs:
      cleantext = p.get_text()
      place_entity = locationtagger.find_locations(text = cleantext)
      region = place_entity.regions
      city = place_entity.cities
      
      if len(region) and len(city) != 0:
        print(f'Year: {year}', "\n")
        print(f'Region Name: {region}', "\n")
        print(f'City Name: {city}', "\n")
        
    
    
```

```python
text_scraper(urls)
```
